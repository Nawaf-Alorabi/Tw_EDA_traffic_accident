# Data Cleaning

```python
# show row that have null
print(df[df.isnull().any(axis=1)])

# how many null
print(df.isnull().sum())

# Drop null
df.dropna(inplace=True)
```

```python
df.columns = [
    'month',         # الشهر
    'year',          # سنة
    'city',          # المدينة
    'males',         # ذكور
    'females',       # أناث
    'inside_city',   # داخل المدينه
    'outside_city',  # خارج المدينه
    'age_under18',   # أقل من 18
    'age_18_30',     # من18 إلى 30
    'age_30_40',     # من 30 على 40
    'age_40_50',     # من40إلى50
    'age_50plus',    # من50فأكثر
    'saudi',         # سعودى
    'non_saudi',     # غير سعودى
    'injuries',      # المصابين
    'deaths',        # المتوفين
]

print('Columns renamed')
df.columns.tolist()
```

```python
df['city'] = df['city'].replace({
    'الرياض'          : 'Riyadh',
    'جده'             : 'Jeddah',
    'المدينه المنوره' : 'Madinah',
    'الشرقيه'         : 'Eastern Region',
    'الحدود الشماليه' : 'Northern Borders',
    'تبوك'            : 'Tabuk',
    'الجوف'           : 'Al-Jouf',
    'حائل'            : 'Hail',
    'نجران'           : 'Najran',
    'القصيم'          : 'Al-Qassim',
    'الباحه'          : 'Al-Baha',
    'عسير'            : 'Asir',
    'جازان'           : 'Jazan',
    'الطائف'          : 'Taif',
    'العاصمه'         : 'Makkah',
    'القريات'         : 'Al-Qurayyat',
})

print(df['city'].value_counts())
```

```python
df['total_accidents'] = df['males'] + df['females']
```
